```python
df2 = pd.DataFrame(columns=('id','createtime','text'))
df3 = pd.read_csv('datanew.csv')
```

```python

def search():
    global df2 
    global df3
    res2 = api.search_tweets(q="cyberattacks  -is:retweet",count=100,tweet_mode='extended')
    for tweet in res2:
        #print(tweet._json['id'])
        status = api.get_status(tweet._json['id'], tweet_mode="extended")
        try: 
            # Is a Retweet
            
            #print(status._json['id'])
            #print(status._json['created_at'])
            #print(status.retweeted_status.full_text)
            df2.loc[df2.index.size] = [status._json['id'],status._json['created_at'], status.retweeted_status.full_text ]
            
        except AttributeError:  # Not a Retweet
            #print(status.full_text)
            df2.loc[df2.index.size] = [status._json['id'],status._json['created_at'], status.full_text]

    df3 = pd.concat([df3,df2],ignore_index=True).drop_duplicates(subset=['id'])
```

```python
import time


def printFun():
    search()
    print ("Number of existing tweets ", len(df3))
    print("The current timestamp is", time.asctime( time.localtime(time.time()) ))

def loop_func(func, second):
 # Executes the func function every second  
    while True:
        func()
        time.sleep(second)
    
loop_func(printFun, 180)
```
